```python
# Step 1: Create an empty dictionary called medical_costs
medical_costs = {}

# Step 2: Add key-value pairs to medical_costs
medical_costs["Marina"] = 6607.0
medical_costs["Vinay"] = 3225.0

# Step 3: Add three more patients to medical_costs
medical_costs.update({"Connie": 8886.0, "Isaac": 16444.0, "Valentina": 6420.0})

# Step 4: Print medical_costs
print("Medical Costs:", medical_costs)

# Step 5: Update Vinay's insurance cost
medical_costs["Vinay"] = 3325.0

# Step 6: Calculate average medical cost
total_cost = sum(medical_costs.values())

# Step 7: Calculate and print average_cost
average_cost = total_cost / len(medical_costs)
print("Average Insurance Cost:", average_cost)

# Step 8: Create lists of names and ages
names = ["Marina", "Vinay", "Connie", "Isaac", "Valentina"]
ages = [27, 24, 43, 35, 52]

# Step 9: Zip names and ages
zipped_ages = zip(names, ages)

# Step 10: Create names_to_ages dictionary using list comprehension
names_to_ages = {name: age for name, age in zipped_ages}

# Step 11: Get Marina's age using .get()
marina_age = names_to_ages.get("Marina", None)
print("Marina's age is", marina_age)

# Step 12: Create an empty dictionary called medical_records
medical_records = {}

# Step 13: Add Marina's medical data to medical_records
medical_records["Marina"] = {
    "Age": 27,
    "Sex": "Female",
    "BMI": 31.1,
    "Children": 2,
    "Smoker": "Non-smoker",
    "Insurance_cost": 6607.0
}

# Step 14: Add other individuals' medical data to medical_records
medical_records.update({
    "Vinay": {
        "Age": 24,
        "Sex": "Male",
        "BMI": 26.9,
        "Children": 0,
        "Smoker": "Non-smoker",
        "Insurance_cost": 3225.0
    },
    "Connie": {
        "Age": 43,
        "Sex": "Female",
        "BMI": 25.3,
        "Children": 3,
        "Smoker": "Non-smoker",
        "Insurance_cost": 8886.0
    },
    "Isaac": {
        "Age": 35,
        "Sex": "Male",
        "BMI": 20.6,
        "Children": 4,
        "Smoker": "Smoker",
        "Insurance_cost": 16444.0
    },
    "Valentina": {
        "Age": 52,
        "Sex": "Female",
        "BMI": 18.7,
        "Children": 1,
        "Smoker": "Non-smoker",
        "Insurance_cost": 6420.0
    }
})

# Step 15: Print medical_records
print("Medical Records:", medical_records)

# Step 16: Print Connie's insurance cost
print("Connie's insurance cost is", medical_records["Connie"]["Insurance_cost"], "dollars.")

# Step 17: Remove Vinay from medical_records
del medical_records["Vinay"]

# Step 18: Print each patient's medical record
for name, data in medical_records.items():
    print(f"{name} is a {data['Age']} year old {data['Sex']} {data['Smoker']} with a BMI of {data['BMI']} and insurance cost of {data['Insurance_cost']}")

# Extra Step 19: Create a function to update medical_records
def update_medical_records(name, medical_data):
    medical_records[name] = medical_data

# Extra Step 20: Create a new dictionary (example)
other_data = {
    "John": {
        "Age": 30,
        "Sex": "Male",
        "BMI": 24.8,
        "Children": 2,
        "Smoker": "Non-smoker",
        "Insurance_cost": 5000.0
    },
    "Emily": {
        "Age": 45,
        "Sex": "Female",
        "BMI": 29.5,
        "Children": 1,
        "Smoker": "Non-smoker",
        "Insurance_cost": 7500.0
    }
}
```
