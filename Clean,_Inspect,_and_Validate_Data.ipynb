```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 2: Load the data and print the first few rows
diabetes_data = pd.read_csv('diabetes.csv')
print(diabetes_data.head())

# Step 3: Number of columns (features)
num_columns = diabetes_data.shape[1]
print("Number of columns (features):", num_columns)

# Step 4: Number of rows (observations)
num_rows = diabetes_data.shape[0]
print("Number of rows (observations):", num_rows)

# Step 5: Checking for null values
null_counts = diabetes_data.isnull().sum()
print(null_counts)

# Step 6: Summary statistics
summary_stats = diabetes_data.describe()
print(summary_stats)

# Step 7: Replace 0 values with NaN in the mentioned columns
diabetes_data[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']] = diabetes_data[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']].replace(0, np.nan)

# Step 8: Check for missing values again
missing_values = diabetes_data.isnull().sum()
print(missing_values)

# Step 9: Print rows with missing values
rows_with_missing_values = diabetes_data[diabetes_data.isnull().any(axis=1)]
print(rows_with_missing_values)


# Step 10: Data types of each column
data_types = diabetes_data.dtypes
print(data_types)

# Step 11: Outcome column type
unique_outcomes = diabetes_data['Outcome'].unique()
print(unique_outcomes)

# Step 12: Resolving the issue with the Outcome column
# Check unique values in the Outcome column
unique_outcomes = diabetes_data['Outcome'].unique()
print(unique_outcomes)

# If the unique values contain non-numeric or unexpected values, we need to clean and convert the column to int64
# For example, if the column contains 'Yes' and 'No' instead of 0 and 1, we can use the map function to convert them.
# Assuming 1 represents 'Yes' and 0 represents 'No':

outcome_mapping = {'Yes': 1, 'No': 0}
diabetes_data['Outcome'] = diabetes_data['Outcome'].map(outcome_mapping)

# Now, the Outcome column should be of type int64
print(diabetes_data.dtypes)

# Step 13: Extensions
# Use .value_counts() to more fully explore the values in each column

# Example: Get the count of unique values in the Outcome column
outcome_counts = diabetes_data['Outcome'].value_counts()
print(outcome_counts)

# Calculate the median of each column
column_medians = diabetes_data.median()

# Replace NaN values with the median of each column
diabetes_data.fillna(column_medians, inplace=True)

```
